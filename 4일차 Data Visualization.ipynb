## How to use Matplotlib

* 기본 구성
    * 데이터 시각화를 수행하면서 기본적으로 작성하는 코드의 구성은 다음과 같다.
        * 절대적인 것은 아니며 개인의 기준
        * 내가 확인할 때도 편하지만, 남들에게 보여줄 때 특히 효과적
```
plt.figure(figsize=(width, height))
plt.title('Matplotlib')
plt.grid()
plt.xticks(iterator)
plt.yticks(iterator)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.plot(data, label='line')
plt.legend()
plt.show()
```

    * figure(figsize=(width, height)): 그래프 크기 지정
        * 항상 먼저 작성, figure에 그래프가 그려짐
    * title: 그래프의 제목
    * xticks, yticks: 눈금 범위 재지정
        * 필요에 따라 사용, 기본적으로 자동 지정됨
    * grid: 격자 표시
        * 파라미터로 False를 넘겨주거나 코드 자체를 지우면 없애기 가능
    * xlabel, ylabel: 축 이름
    * plot: 시각화 함수
        * label은 데이터의 범례를 나타냄
        * 예시는 plot 함수이지만 상황에 따라 적절히 변경
    * legend: 범례 표시
        * 항상 시각화 함수 이후에 작성
    * show: 그래프 표시

````python
import matplotlib.pyplot as plt
import numpy as np
````

### plot

* 데이터의 추세를 확인할 때 주로 사용

````python
data1 = np.random.rand(20) #20개 랜덤값(0-1 사이)
data2 = np.random.rand(20)

plt.figure(figsize=(10, 4)) # 가로 10 세로 4
plt.title('Plot')
plt.grid()
plt.xticks(np.arange(20)) #0-19 까지(numpy 사용 시 시간 감소)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.plot(data1, 'o--', label='plot1') # option 많음
plt.plot(data2, 'x-', label='plot2')
plt.legend()
plt.show()
````

````python
d1 = np.random.rand(10)

plt.figure(figsize=(5,3))
plt.grid()
plt.xticks(np.arange(10))
plt.plot(d1, label='data1')
plt.show()
````

### scatter

* 데이터들의 관계를 분석할 때 주로 사용

````python
x = np.arange(20)
y1 = np.random.rand(20)
y2 = np.random.rand(20)

plt.figure(figsize=(6, 4))
plt.title('Scatter')
plt.grid()
plt.xticks(np.arange(0, 20, 2))
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.scatter(x, y1, label='scatter1')
plt.scatter(x, y2, label='scatter2')
plt.legend()
plt.show()
````

````python
age = [19,19,19,19,19,18,18,18,19,20,20,18,18,20,10]
peopledata = np.arange(15)
agedata = np.array(age)

plt.figure(figsize=(5,2))
plt.grid()
plt.xticks(np.arange(15))
plt.scatter(peopledata,agedata)
plt.show()
````

### hist

* 데이터들의 분포를 확인할 때 주로 사용

````python
y1 = np.array([np.random.randint(10) for _ in range(100)])
y2 = np.array([np.random.randint(10) for _ in range(100)])

plt.figure(figsize=(6, 6))
plt.title('Hist')
plt.grid()
plt.xticks(np.arange(10))
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.hist((y1, y2), label=('Hist1', 'Hist2'))
plt.legend()
plt.show()
````

````python
xt1 = np.array([0,0,0,0,1,1,1,2,2,2,5,5,4,4,3,3,3,5,5,5,5,5,5])

plt.figure(figsize=(8,2))
plt.grid()
plt.xticks(np.arange(10))
plt.hist(xt1)
plt.show()
````

### bar

* 데이터들의 수치를 확인할 때 주로 사용

````python
x = np.arange(10)
y1 = np.array([np.random.randint(10) for _ in range(100)])
y2 = np.array([np.random.randint(10) for _ in range(100)])
w = 0.35

plt.figure(figsize=(6, 6))
plt.title('Bar')
plt.grid(axis='y')
plt.xticks(x+w/2, np.arange(10))
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.bar(x, np.bincount(y1), width=w, label='Bar1')
plt.bar(x+w, np.bincount(y2), width=w, label='Bar2')
plt.legend()
plt.show()
````

### pie

* 데이터들의 비율을 확인할 때 주로 사용

````python
x = np.arange(10)
y = np.array([np.random.randint(10) for _ in range(100)])

plt.figure(figsize=(6, 6))
plt.pie(np.bincount(y), labels=np.arange(10), startangle=90,  autopct='%.1f%%')
plt.show()
````

````python
x = np.arange(10)
y = np.array([np.random.randint(10) for _ in range(100)])

plt.figure(figsize=(4,4))
plt.pie(np.bincount(y), labels=np.arange(10), startangle=90,  autopct='%.1f%%')
plt.show()
````

### boxplot

````python
x = np.arange(13)
xt2 = np.array([4,9,1,9,8,8,3,11,3,3,5,9,8])

plt.figure(figsize=(4,4))
plt.pie(np.bincount(xt2), labels=np.bincount(xt2) ,startangle=30, autopct='%.1f%%')
plt.show()
````

* 데이터에 포함된 이상치를 확인할 때 주로 사용

````python
data = np.random.rand(20, 4)
row = np.random.choice(20, 8, replace = False)

for c in np.random.choice(4, 3):
    data[np.random.choice(row, 15), c] *= 3

plt.figure(figsize=(10,6))
plt.boxplot(data)
plt.xticks(range(1, 5), ['A', 'B', 'C', 'D'])
plt.show()
````

````python
d1 = np.array([2.0,1.5,0.6,1.5,0.1,1.0,-1.4,-2.5,-2.0,0.7,0.5])
plt.figure(figsize=(5,4))
plt.boxplot(d1)
plt.show()
````

## How to use Seaborn

* seaborn은 시각화뿐만 아니라 학습을 위한 데이터들을 제공

````python
import seaborn as sns
import pandas as pd
````

````python
df = sns.load_dataset("titanic")
df
````

### plot

````python
plt.title('an age-based fare')
plt.grid()
sns.lineplot(data=df, x='age', y='fare', hue='sex', style='sex', markers= ["o","X"], ci=None)
plt.show()
````

### scatterplot

* 데이터들의 관계를 분석할 때 주로 사용

````python
plt.title('scatterplot')
sns.scatterplot(data=df, x='age', y='fare', hue='survived')

# target = 예측하고 싶은 값 / feature = target 에 영향을 주는 값
plt.show()
````

### histplot

* 데이터들의 분포를 확인할 때 주로 사용

````python
plt.title('hisplot')
sns.histplot(data=df, x='age', bins=16, hue='survived')
plt.show()
````

### kdeplot

* 데이터들의 분포 비율을 확인할 때 주로 사용

#### Univariate

````python
sns.kdeplot(data=df, x='age', hue='survived')
plt.show()
````

#### Bivariate

````python
sns.kdeplot(data=df, x='age', y='fare', hue='survived')
plt.show()
````

### countplot

* 데이터들의 수치를 확인할 때 주로 사용

````python
plt.title('countplot')
sns.countplot(data=df, x='sex', hue='survived')
plt.show()
````

### boxplot

* 데이터에 포함된 이상치를 확인할 때 주로 사용

````python
plt.title('boxplot')
sns.boxplot(data=df)
plt.show()
````

### pairplot

* 다차원 데이터들의 관계를 분석할 때 주로 사용

````python
sns.pairplot(data=sns.load_dataset("iris"), hue='species')
plt.show()
````

### heatmap

* 2차원 데이터들의 관계를 분석할 때 주로 사용

````python
plt.title('heatmap')
sns.heatmap(data=df.corr(), annot=True, fmt=".2f")
plt.show()
````
